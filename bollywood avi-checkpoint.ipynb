```python
import pandas as pd
import matplotlib as plt
import numpy as np

df = pd.read_csv("D:\\FML-AVI-230957164\\Week 1 M&C\\bollywood.csv")
df.head()
```

```python
# Find the number of records in the dataset
record_count = df.shape[0]
print(f"Number of records: {record_count}")
```

```python
# Count the number of movies in each genre
genre_counts = df['Genre'].value_counts()

# Sort the genre releases in descending order
genre_counts_sorted = genre_counts.sort_values(ascending=False)
print(genre_counts_sorted)
```

```python
# Get the genre with the highest number of releases
top_genre = genre_counts_sorted.idxmax()
print(f"The genre with the highest number of releases is: {top_genre}")
```

```python
# Cross-tabulate Genre and ReleaseTime
genre_release_time = pd.crosstab(df['Genre'], df['ReleaseTime'])
print(genre_release_time)
```

```python
# Convert ReleaseDate to datetime format
df['Release Date'] = pd.to_datetime(df['Release Date'], errors='coerce')

# Extract the month from ReleaseDate
df['Month'] = df['Release Date'].dt.month

# Count the number of releases per month
month_counts = df['Month'].value_counts().sort_index()

print(f"Movies released by month: \n{month_counts}")
```

```python
# Filter movies with budget >= 25 crore
high_budget_movies = df[df['Budget'] >= 25]

# Count the number of high-budget movie releases per month
high_budget_month_counts = high_budget_movies['Month'].value_counts().sort_index()

print(f"High-budget movie releases by month: \n{high_budget_month_counts}")
```

```python
# Calculate ROI
df['ROI'] = (df['BoxOfficeCollection'] - df['Budget']) / df['Budget']

# Sort by ROI and get the top 10 movies
top_10_roi_movies = df[['MovieName', 'ROI']].sort_values(by='ROI', ascending=False).head(10)

print("Top 10 movies with maximum ROI:\n", top_10_roi_movies)
```

```python
# Group by ReleaseTime and calculate the average ROI
avg_roi_release_time = df.groupby('ReleaseTime')['ROI'].mean()

print(f"Average ROI for different release times:\n{avg_roi_release_time}")
```

```python
# Calculate the correlation between BoxOfficeCollection and YoutubeLikes
correlation = df['BoxOfficeCollection'].corr(df['YoutubeLikes'])
print(f"Correlation between BoxOfficeCollection and YoutubeLikes: {correlation}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Boxplot for YouTube likes by genre
plt.figure(figsize=(10, 6))
sns.boxplot(x='Genre', y='YoutubeLikes', data=df)
plt.title('YouTube Likes Distribution by Genre')
plt.xticks(rotation=45)
plt.show()
```

```python
# Select the relevant columns for correlation
correlation_matrix = df[['Budget', 'BoxOfficeCollection', 'YoutubeViews', 'YoutubeLikes', 'YoutubeDislikes']].corr()

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap for Budget, BoxOfficeCollection, Youtube Views/Likes/Dislikes')
plt.show()

# Alternatively, you can use a pairplot for visualizing pairwise relationships
sns.pairplot(df[['Budget', 'BoxOfficeCollection', 'YoutubeViews', 'YoutubeLikes', 'YoutubeDislikes']])
plt.show()
```

```python
# Filter movies released in 2013-2015
df['Year'] = df['Release Date'].dt.year
df_filtered = df[(df['Year'] >= 2013) & (df['Year'] <= 2015)]

# Scatter plot for Genre vs BoxOfficeCollection
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Genre', y='BoxOfficeCollection', data=df_filtered)
plt.title('Genre vs Box Office Collection (2013-2015)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Scatterplot for Budget vs BoxOfficeCollection by Genre
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Budget', y='BoxOfficeCollection', hue='Genre', data=df)
plt.title('Budget vs Box Office Collection by Genre')
plt.show()
```

```python
# Boxplot for Budget distribution by Genre
plt.figure(figsize=(10, 6))
sns.boxplot(x='Genre', y='Budget', data=df)
plt.title('Budget Distribution by Genre')
plt.xticks(rotation=45)
plt.show()
```

```python
# Group by the 'Year' and get the count of movies for each year
movies_per_year = df['Year'].value_counts()

# Convert the result into a DataFrame and rename the column to 'count'
movies_per_year_df = movies_per_year.reset_index()
movies_per_year_df.columns = ['Year', 'count']

# Plotting the best fit graph
sns.lmplot(x='Year', y='count', data=movies_per_year_df, aspect=2, height=6)
plt.title('Movies Released per Year (2013-2015)')
plt.show()
```
